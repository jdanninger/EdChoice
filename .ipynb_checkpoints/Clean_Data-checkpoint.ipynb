## Year 2011 - 2012

```python
# Read data with correct types
dtype_specification = {'Building IRN': str, 'District IRN': str, "Street address":str, 'City and Zip code':str} 
AY2012 = pd.read_excel('./2011-2012/1112_LRC_BUILDING.xls', dtype=dtype_specification)

# Drop bad rows 
AY2012 = AY2012.dropna(subset=[AY2012.columns[0]])
AY2012 = AY2012.dropna(subset=[AY2012.columns[1]])
AY2012 = AY2012.dropna(subset=[AY2012.columns[2]])
AY2012 = AY2012.dropna(subset=[AY2012.columns[3]])


def clean2012(row):
    new_row = ({
        'District Name' : row[3],
        'District IRN' : row[2],
        'Building Name' : row[1],
        'Building IRN' : row[0],
        'Building Adress' : str(row[6])+ " "+str(row[7]),
        'Year':2012,
        'Preformance Index':row[94],
        'School Type': row[12],
    })
    final_output.loc[len(final_output)] = new_row
##    print(new_row)
    return row


AY2012 = AY2012.apply(clean2012, axis=1)
final_output
```
