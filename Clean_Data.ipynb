## Read Data

## Year 2008 - 2009

```python
# Read data with correct types
dtype_specification = {'Building IRN': str, 'District IRN': str, "Street address":str, 'City and Zip code':str} 
AY2009 = pd.read_excel('./2008-2009/0809_LRC_BUILDING.xls', dtype=dtype_specification)

# Drop bad rows 
AY2009 = AY2009.dropna(subset=[AY2009.columns[0]])
AY2009 = AY2009.dropna(subset=[AY2009.columns[1]])
AY2009 = AY2009.dropna(subset=[AY2009.columns[2]])
AY2009 = AY2009.dropna(subset=[AY2009.columns[3]])


def clean2009(row):
    new_row = ({
        'District Name' : row[3],
        'District IRN' : row[2],
        'Building Name' : row[1],
        'Building IRN' : row[0],
        'Building Adress' : str(row[6])+ " "+str(row[7]),
        'Year':2009,
        'Preformance Index':row[107],
        'School Type': row[10],
        'AYP Met?' : row[110]
    })
    final_output.loc[len(final_output)] = new_row
##    print(new_row)
    return row


AY2009 = AY2009.apply(clean2009, axis=1)
final_output
```
