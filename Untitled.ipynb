## Dependencies

```python
import pandas as pd
```

## Read Data

```python
prof = pd.read_excel('AMO_BLDG.xls', sheet_name = 'AMO Proficiency' )
prof = prof.drop(prof.index[0])
curr = 1
prof
```

## Get Profciency English and Math

```python
while curr < prof.shape[0]:
    prof = prof.drop(prof.index[(curr) : (curr + 9)])
    curr = curr +1


cols_to_drop = prof.columns.tolist()
cols_to_drop = cols_to_drop[6:]
cols_to_drop.remove('Current Year Proficient % Reading')
cols_to_drop.remove('Previous Year Proficient % ')
cols_to_drop.remove('Current Year Proficient % .1')
cols_to_drop.remove('Previous Year Proficient % .1')
prof = prof.drop(cols_to_drop, axis=1)

new_name = {"Current Year Proficient % " : "Current Year Reading Proficient %", "Previous Year Proficient % " : "Previous Year Reading Proficient %",}
prof = prof.rename(columns = new_name)

new_name = {"Current Year Proficient % .1" : "Current Year Math Proficient %", "Previous Year Proficient % .1" : "Previous Year Math Proficient %",}
prof = prof.rename(columns = new_name)

print(prof.columns.tolist())
prof

```
